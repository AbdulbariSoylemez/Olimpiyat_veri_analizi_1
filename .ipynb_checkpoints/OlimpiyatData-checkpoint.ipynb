```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### kodumuza uyarı çıktığında uyarıyı görmezden gelelim 

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv("athlete_events.csv")
data.head()
```

#### şimdi değişkenlerimizin isimlerini değiştirelim 

```python
isimler = {
    "ID": "Kimlik",
    "Name": "İsim",
    "Sex": "Cinsiyet",
    "Age": "Yaş",
    "Height": "Boy",
    "Weight": "Kilo",
    "Team": "Takım",
    "NOC": "NOC",
    "Games": "Oyunlar",
    "Year": "Yıl",
    "Season": "Mevsim",
    "City": "Şehir",
    "Sport": "Spor",
    "Event": "Etkinlik",
    "Medal": "Madalya"
}

data.rename(columns=isimler, inplace=True)

print(data.columns)
```

#### şimdi eksik verilerimize bakalım var işe bunları yerlerini dolduralım 

```python
data.isnull().sum()
```

#### şmdi eksik verilerin yerine ortalama değerlerini girelim 

```python
veri=data.copy()   #  bazı yerlerde asıl olan  veriye ihtiyaç olacaktır ondan copyasını alalım 
veri["Yaş"]=veri["Yaş"].fillna(data.Yaş.mean())   # yaş verimizin eksik değerlerine ortalama değerler ile değiştirdik 
print(" eksik verilere ortalama değerleri ekledikten sonra veri.Yaş değikeninde eksik değer kalmıyor data \n ve ortalamaları aynı oluyor" , data.Yaş.mean()," and ",veri.Yaş.mean())
```

```python
veri.isnull().sum()
```

#### şimdi ise eksik kalan Boy ve Kilo değeri içinde eksik edeğerlere ortalama değer ekleme yapalım 

```python
veri.Boy=veri.Boy.fillna(data.Boy.mean()) # boy değeri için eksik değerler ortalma değer ile dolduruldu 
veri.Kilo=veri.Kilo.fillna(data.Kilo.mean()) # Kilo değeri için eksik değerler ortalama değer ile dolduruldu
```

```python
veri.isnull().sum()
print(veri.Boy.mean(),
data.Boy.mean())
print( veri.Kilo.mean(),data.Kilo.mean())
```

```python
veri.head()
```

##### şimdi ise veride ihtiyaç olmuyan kolonları kaldıralım bunlar bizim Oyunlar ve İsim değikenleridir

###### şimdi ile histogram grafiğini çizelim 

```python
def histogram(değişken):
    plt.figure()
    plt.hist(veri[değişken],bins=60,color="orange") # bins : aralık sayısını sayısını belirler 
    plt.title("veri sıklığı - {}".format(değişken))
    plt.xlabel(değişken)
    plt.xlabel("Frekans")
    plt.show()
```

```python
değişkenler=["Yaş","Boy","Kilo","Yıl"]
for i in değişkenler:
    histogram(i)
```

```python
plt.boxplot(veri.Yaş)
plt.title("Madalya oranı ")
plt.ylabel("Yaş frenkansı")
plt.xlabel("Madalya")
plt.show()
```

##### kategorik değerlerde ilk 5 veriyi gösterelim 

```python
def kategor(değişken, n):
    dff=veri[değişken]
    dfx=dff.value_counts()
    dfx=dfx[:n]
    plt.figure()
    plt.bar(dfx.index,dfx,color="orange")
    plt.xticks(dfx.index,dfx.index.values)
    plt.xticks(rotation=45)
    plt.title(değişken+" frenkansı")
    print(dfx)
    plt.show()
    
```

```python
değişkenler = ['Şehir', 'Spor', 'Etkinlik', 'Madalya',"İsim"]
for değişken in değişkenler:
    kategor(değişken, 5)
```

```python
def madyaş(A,B):
    verisayısıA=veri[A].value_counts()
    verisayısıA=verisayısıA[:10]
    verisayısıB=veri[B].value_counts()
    plt.figure()
    plt.scatter(verisayısıA.index,verisayısıA,color="red")
    plt.xticks(rotation=45)
    plt.show()
```

```python
madyaş("Yaş","Madalya")
```

```python
geçiciveri=veri.copy()
geçiciveri=pd.get_dummies(geçiciveri,columns=["Madalya"])
geçiciveri.head(3)
```

##### corr değerine bakalım madalyanın yaşlara göre corelasyonu 

```python
geçiciveri.loc[:,["Yaş","Madalya_Bronze","Madalya_Gold","Madalya_Silver"]].corr()
```

```python
takımlar=geçiciveri["Takım"].value_counts()
takımlar=takımlar[:10]
plt.figure()
plt.bar(takımlar.index,takımlar,color="red")
plt.xlabel("Takımlar")
plt.xticks(rotation=45)
plt.ylabel("madalya sayısı")
plt.title("madalyaların takımlara dağılışı")
plt.show()
```

```python
geçiciveri.loc[:,["Takım","Madalya_Bronze","Madalya_Gold","Madalya_Silver"]].groupby(["Takım"],
                           as_index=False).sum().sort_values(by="Madalya_Gold",
                           ascending=False)[:10]
```
